fazer um programa que converte a temperatura de celsius para fahrenheit.

entrada: graus celsius

processamentoi: converter a temperatura para fahrenheit

saida: mostrar a temperatura em fahrenheit

F = (0 °C × 9/5) + 32

```python
# entrada

C = int (input("digite os garus celsius: "))

# processamento

 F = (C * 9/5) + 32

# saida

print (f"temperatura em fahrenheit: {F}")
```

```python
# operadores aritimericos

# entrada

a = int (input ("digite o primeiro numero: "))
b = int (input ("digite o segundo numero: "))

# processamento

# soma de dois numeros e subtração de dois numeros

soma = a + b
subtração = a-b
multiplicacao = a*b
divisao = a/b
div_inteira = a//b
resto = a%b  # resto da divisao


# saida

print (f"a soma de {a} mais {b} é  igual a: {soma}")
print (f"a diferença de {a} menos {b} é  igual a: {subtração}")
print (f"a multiplicacao de {a} vezes {b} é  igual a: {multiplicacao}")
print (f"a divisao de {a} dividido {b} é  igual a: {divisao}")
print (f"a divisao inteira de {a} dividido {b} é  igual a: {div_inteira}")
print (f"o resto da divisao de {a} dividido {b} é  igual a: {resto}")

```
