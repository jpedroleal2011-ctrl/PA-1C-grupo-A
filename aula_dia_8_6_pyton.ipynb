```python
#4 faça um programa que calcule a soma dos numeros digitados pelo usuario
quantidade = int(input("Quantos números você deseja somar? "))

soma = 0

for i in range(quantidade):
    numero = float(input(f"Digite o {i+1}º número: "))
    soma += numero

    print(f"\nA soma dos {quantidade} números digitados é: {soma}")
```

```python
#5 solicite um numero de 1 a 10 ao usuario e exiba a tabuada
numero = int(input("Digite um número de 1 a 10 para ver a tabuada: "))

if 1 <= numero <= 10:
    print(f"Tabuada do {numero}:")
    for i in range(1, 11):
        print(f"{numero} x {i} = {numero * i}")
else:
    print("digite um número entre 1 e 10.")
```

```python
#6 solicite ao usuario 4 notas e calcule a média
notas = []
soma_notas = 0

for i in range(4):
    while True:
        try:
            nota = float(input(f"Digite a {i+1}ª nota: "))
            if 0 <= nota <= 10:
                notas.append(nota)
                soma_notas += nota
                break
            else:
                print("A nota deve ser entre 0 e 10.")
        except ValueError:
            print("Entrada inválida. Digite um número para a nota.")

media = soma_notas / len(notas)
print(f"As notas digitadas foram: {notas}")
print(f"A média das notas é: {media:.2f}")
```

```python
#7 crie um programa que solicite uma senha até que o usuario digite a senha correta

senha_correta = "minhasenha123"
senha_digitada = ""

while senha_digitada != senha_correta:
    senha_digitada = input("Digite a senha: ")
    if senha_digitada == senha_correta:
        print("Senha correta! Acesso concedido.")
    else:
        print("Senha incorreta. Tente novamente.")
```

```python
#8 faça um programa que receba varios valores e quantidade de produtos e no final exibe o valor total da compra

total_compra = 0

while True:
    try:
        preco_produto = float(input("Digite o preço do produto (ou 0 para finalizar): "))
        if preco_produto == 0:
            break

        quantidade_produto = int(input("Digite a quantidade do produto: "))

        total_compra += (preco_produto * quantidade_produto)

    except ValueError:
        print("Entrada inválida. Por favor, digite um número para o preço e a quantidade.")

print(f"O valor total da compra é: R$ {total_compra:.2f}")
```

```python
#9 peça ao usuario para digitar 5 numeros e no final exiba uma mensagem com o maior numero

maior_numero = float('-inf') # Inicializa com o menor número possível

for i in range(5):
    while True:
        try:
            numero = float(input(f"Digite o {i+1}º número: "))
            if numero > maior_numero:
                maior_numero = numero
            break
        except ValueError:
            print("Entrada inválida. Por favor, digite um número.")

print(f"O maior número digitado foi: {maior_numero}")
```

```python
#10 crie um jogo de adivinhação, o usuario tenta adivinhar um número secreto

import random

numero_secreto = random.randint(1, 999)
tentativas = 0
max_tentativas = 3

print("Bem-vindo ao jogo de adivinhação!")
print(f"Tente adivinhar um número entre 1 e 10. Você tem {max_tentativas} tentativas.")

while tentativas < max_tentativas:
    try:
        palpite = int(input("Digite seu palpite: "))
        tentativas += 1

        if palpite == numero_secreto:
            print("Você acertou!")
            break
        elif palpite < numero_secreto:
            print("O número secreto é maior.")
        else:
            print("O número secreto é menor.")

    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro.")

if palpite != numero_secreto:
    print(f"Suas {max_tentativas} tentativas acabaram. O número secreto era {numero_secreto}.")
```
