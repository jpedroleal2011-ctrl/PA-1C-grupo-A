```python
#  escrever um programa que o usuario digiota um numero e o programa imprima de 0 ate o numero;
numero = int(input("Digite um número inteiro: "))

for i in range(numero + 1):
    print(i)
```

```python
# escrever um programa que o usuario digita um numero eo programa imprimi em ordem decrerscente
# ate zero;
numero = int(input("Digite um número inteiro: "))

for i in range(numero, -1, -1):
    print(i)
```

```python
# peça um numero e mostre todos os numeros pares ate o numero;
numero = int(input("Digite um número inteiro: "))

for i in range(numero + 1):
    if i % 2 == 0:
        print(i)
```

```python
# peça um numero e calcule a soma de 1 ate o numero;
numero = int(input("Digite um número inteiro: "))

soma = 0
for i in range(1, numero + 1):
    soma += i

print(f"A soma de 1 até {numero} é: {soma}")
```

```python
# peça uma palavra e conte as letras: EX. python - a palavra tem 6 letras
palavra = input("Digite uma palavra: ")
numero_letras = len(palavra)
print(f"A palavra '{palavra}' tem {numero_letras} letras.")
```

```python
# peca uma palavra e conte as vogais;
palavra = input("Digite uma palavra: ").lower()
vogais = "aeiou"
contador_vogais = 0

for letra in palavra:
    if letra in vogais:
        contador_vogais += 1

print(f"A palavra '{palavra}' tem {contador_vogais} vogais.")
```

```python
# peça ao usuario para digitar 05 numeros e informe o valor
numeros = []
for i in range(5):
    while True:
        try:
            numero = float(input(f"Digite o {i+1}º número: "))
            numeros.append(numero)
            break
        except ValueError:
            print("Entrada inválida. Por favor, digite um número.")

print("Os números digitados foram:")
for num in numeros:
    print(num)
```
