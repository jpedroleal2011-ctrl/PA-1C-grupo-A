```python
#Solicite ao usuário um número inteiro positivo.
#Utilizando um laço de repetição, calcule e exiba a soma de todos os números de 1 até o número informado.
numero = int(input("Digite um número inteiro: "))

soma = 0
for i in range(1, numero + 1):
    soma += i

print(f"A soma de 1 até {numero} é: {soma}")
```

```python
#Solicite ao usuário um número inteiro e exiba sua tabuada de 1 até 10.

numero = int(input("Digite um número de 1 a 10 para ver a tabuada: "))

if 1 <= numero <= 10:
    print(f"Tabuada do {numero}:")
    for i in range(1, 11):
        print(f"{numero} x {i} = {numero * i}")
else:
    print("digite um número entre 1 e 10.")
```

```python
#Solicite ao usuário um número inteiro positivo e calcule seu fatorial utilizando um laço de repetição.


num = int(input("Digite um numero: "))

fatorial = 1
for i in range(1, num + 1):
    fatorial *= i

print(f"A fatorial de {num} é: {fatorial}")
```

```python
#Um caixa eletrônico possui notas de:
# Solicite ao usuário o valor do saque e informe a menor quantidade possível de notas para realizar o pagamento.

saque = int(input("Digite o valor do saque: "))

notas = [100, 50, 20, 10, 5, 2, 1]
quantidade_notas = {}

for nota in notas:
    if saque >= nota:
        quantidade = saque // nota
        quantidade_notas[nota] = quantidade
        saque %= nota

for nota, quantidade in quantidade_notas.items():
    if quantidade > 0:
        print(f"{quantidade} nota(s) de R${nota}")



```
