```python
# escrever um programa que faça as quatro operaçoes aritmeticas (calculadora simples)

print ("-------calculadora simples-------")
print ("-------------menu----------------")
print ("                                 ")
print ("soma = 1")
print ("                                 ")
print ("subtração = 2")
print ("                                 ")
print ("multiplicação = 3")
print ("                                 ")
print ("divisão = 4")
print ("---------------------------------")

#entrada

numero = int(input("ecolha a operação que deseja realizar: "))
print("")
a=int(input("digite o primeiro numero: "))
b=int(input("digite o segundo numero: "))
print("")

if numero == 1:
    print("soma: ", a+b)
if numero == 2:
    print("subtração: ", a-b)
if numero == 3:
    print("multiplicação: ", a*b)
if numero == 4:
    print("divisão: ", a/b)
```

```python
# faca um programa que classifique triangulos

#entrada

a=int(input("digite o primeiro lado: "))
b=int(input("digite o segundo lado: "))
c=int(input("digite o terceiro lado: "))
print("")

if a+b>c and a+c>b and b+c>a:
    print("é um triangulo")

if a+b<c or a+c<b or b+c<a:
    print("não é um triangulo")

if a==b==c:
    print("equilátero")

elif a==b and a==c and b==c:
    print("isósceles")

else:
    print("escaleno")
```
