```python
# 1 - Escrever um programa qu eo usuario digita um numero e o programa imprimi ate o numero digitado
num = int(input("Digite um numero: "))

while num >= 1:
    print(num)
    num -= 1
```

```python
# 2 - Escrever um programa que o usuario digita um numero e o programa imprimi o numero digitado ate o zero (contagem regressiva)
num = int(input("Digite um numero para a contagem regressiva: "))

while num >= 0:
    print(num)
    num -= 1
```

```python
# 3 - Escrever um programa que o usuario digita um numero e o programa imprimi numeros pares
num = int(input("Digite um numero: "))

print(f"Numeros pares ate {num}:")
for i in range(num + 1):
    if i % 2 == 0:
        print(i)
```

```python
# 4 - Escrever um programa que o usuario digita um numero e o programa imprime numeros impares
num = int(input("Digite um numero: "))

print(f"Numeros impares ate {num}:")
for i in range(num + 1):
    if i % 2 != 0:
        print(i)
```

```python
# 5 - escrever um programa que o usuario digita um numero e o programa soma os numeros ate o numero digitado
num = int(input("Digite um numero: "))

sum_numbers = 0
for i in range(1, num + 1):
    sum_numbers += i

print(f"A soma dos numeros ate {num} eh: {sum_numbers}")
```

```python
# 6 - escrever um programa que o usuario digita uma palavra e o programa conta as letras da palavra
palavra = input("Digite uma palavra: ")

numero_letras = len(palavra)

print(f"A palavra '{palavra}' tem {numero_letras} letras.")
```

```python
# 8 - escrever um programa que o usuario digita uma palavra e o programa  imprimi a palavra invertida
palavra = input("Digite uma palavra: ")

palavra_invertida = palavra[::-1]

print(f"A palavra invertida eh: {palavra_invertida}")
```

```python
# 9 - escrever um programa que o usuario digita 04 numeros e o programa imprimi o maior

numeros = []

for i in range(4):
    numero = int(input(f"Digite o {i + 1} numero: "))
    numeros.append(numero)

maior = max(numeros)

print(f"O maior numero eh: {maior}")
```

```python
# 10 - piramide numerica simetrica
# faca um programa que receba um numero N e imprimi  uma piramede perfeitamente centralizada
n = int(input("Digite um numero para a piramide (N): "))

for i in range(1, n + 1):
    # Imprime os espacos iniciais para centralizar a piramide
    print(" " * (n - i), end="")

    # Imprime a parte crescente dos numeros
    for j in range(1, i + 1):
        print(j, end="")

    # Imprime a parte decrescente dos numeros (se houver)
    for k in range(i - 1, 0, -1):
        print(k, end="")

    # Quebra de linha para a proxima fila
    print()
```
